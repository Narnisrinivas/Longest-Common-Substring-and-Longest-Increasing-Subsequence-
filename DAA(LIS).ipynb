```python
# LIS Brute Force

import pandas as pd
import itertools
import time


def is_increasing(seq):
    return all(seq[i] < seq[i + 1] for i in range(len(seq) - 1))


def lis_bruteforce(arr):
    n = len(arr)
    max_len = 0
    best_seq = []

    for r in range(1, n + 1):
        for subseq in itertools.combinations(arr, r):
            if is_increasing(subseq):
                if len(subseq) > max_len:
                    max_len = len(subseq)
                    best_seq = list(subseq)

    return max_len, best_seq


# Read Excel file
file_name = r"C:\Users\narni\OneDrive\Desktop\sem-4 projects\daa\input_10.xlsx"   # SMALL dataset only
data = pd.read_excel(file_name)

arr = data.iloc[:, 0].tolist()

start = time.time()
length, subsequence = lis_bruteforce(arr)
end = time.time()

print("LIS Length (Brute Force):", length)
print("LIS Subsequence:", subsequence)
print("Time taken:", end - start, "seconds")
```

```python
# LIS DP

import pandas as pd
import time


def lis_dp(arr):
    n = len(arr)

    # L[i] = length of LIS ending at index i
    L = [1] * n

    # prev[i] = previous index in LIS
    prev = [-1] * n

    for j in range(n):
        for i in range(j):
            if arr[i] < arr[j] and L[i] + 1 > L[j]:
                L[j] = L[i] + 1
                prev[j] = i

    # Find maximum length and its index
    max_len = max(L)
    idx = L.index(max_len)

    # Reconstruct LIS
    lis = []
    while idx != -1:
        lis.append(arr[idx])
        idx = prev[idx]

    lis.reverse()

    return max_len, lis


# Read Excel file
file_name = r"C:\Users\narni\OneDrive\Desktop\sem-4 projects\daa\input_10.xlsx"
data = pd.read_excel(file_name)

# Handle possible empty values
arr = data.iloc[:, 0].dropna().tolist()

start = time.time()
length, subsequence = lis_dp(arr)
end = time.time()

print("LIS Length (DP):", length)
print("LIS Subsequence:", subsequence)
print("Time taken:", end - start, "seconds")
```

```python
# LIS Memo


import pandas as pd
import time
import sys

# Increase recursion limit (important for larger inputs)
sys.setrecursionlimit(10**7)


def lis_memo(arr):
    n = len(arr)

    # memo[i] stores LIS ending at index i
    memo = [-1] * n

    def lis_ending_at(i):
        # Base case
        if i == 0:
            return 1

        # If already computed
        if memo[i] != -1:
            return memo[i]

        best = 1  # minimum LIS length

        for j in range(i):
            if arr[j] < arr[i]:
                best = max(best, lis_ending_at(j) + 1)

        memo[i] = best
        return best

    ans = 1
    for i in range(n):
        ans = max(ans, lis_ending_at(i))

    return ans


if __name__ == "__main__":

    # Excel file path (CHANGE THIS)
    file_path =  r"C:\Users\narni\OneDrive\Desktop\sem-4 projects\daa\input_10.xlsx"

    # Read Excel file
    df = pd.read_excel(file_path)

    # Convert first column to list (remove empty values)
    arr = df.iloc[:, 0].dropna().tolist()

    # Measure time
    start = time.time()
    result = lis_memo(arr)
    end = time.time()

    print("Input Array:", arr)
    print("LIS Length (Memoization):", result)
    print("Time Taken:", end - start, "seconds")
```
