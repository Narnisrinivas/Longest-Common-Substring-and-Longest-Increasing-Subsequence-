```python
# LCS Brute Force

import pandas as pd
import time

def lcs_bruteforce(s1, s2):
    n = len(s1)
    m = len(s2)
    max_len = 0

    for i in range(n):
        for j in range(i + 1, n + 1):
            sub = s1[i:j]
            if sub in s2:
                max_len = max(max_len, len(sub))

    return max_len


if __name__ == "__main__":
    # Read Excel file
    file_path = r"C:\Users\narni\OneDrive\Desktop\sem-4 projects\daa\lcs_50_100.xlsx"  # change file name
    df = pd.read_excel(file_path)

    s1 = str(df.iloc[0, 0])
    s2 = str(df.iloc[1, 0])

    start = time.time()
    length = lcs_bruteforce(s1, s2)
    end = time.time()

    print("Length of Longest Common Substring (Brute Force):", length)
    print("Time taken:", end - start, "seconds")
```

```python
# LCS DP

import pandas as pd
import time

def longest_common_substring(s1, s2):
    n = len(s1)
    m = len(s2)

    dp = [[0 for _ in range(m + 1)] for _ in range(n + 1)]

    max_len = 0
    end_pos = 0

    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i - 1] == s2[j - 1]:   #  fixed index (was s2[i-1])
                dp[i][j] = dp[i - 1][j - 1] + 1

                if dp[i][j] > max_len:
                    max_len = dp[i][j]
                    end_pos = i
            else:
                dp[i][j] = 0

    substring = s1[end_pos - max_len: end_pos]

    return max_len, substring


# Read Excel file
file_name = r"C:\Users\narni\OneDrive\Desktop\sem-4 projects\daa\lcs_50_100.xlsx"
data = pd.read_excel(file_name, header=None)

str1 = str(data.iloc[0, 0])
str2 = str(data.iloc[1, 0])

start = time.time()
length, substring = longest_common_substring(str1, str2)
end = time.time()

print("String 1:", str1)
print("String 2:", str2)
print("Longest Common Substring:", substring)
print("Length:", length)
print("Time taken:", end - start, "seconds")
```

```python
# LCS Memo

import pandas as pd
import time
import sys

def lcs_substring_memo(s1, s2):
    n = len(s1)
    m = len(s2)

    memo = [[-1] * (m + 1) for _ in range(n + 1)]
    max_length = 0
    end_index = 0   # to reconstruct substring

    def solve(i, j):
        nonlocal max_length, end_index

        if i == 0 or j == 0:
            return 0

        if memo[i][j] != -1:
            return memo[i][j]

        if s1[i - 1] == s2[j - 1]:
            memo[i][j] = solve(i - 1, j - 1) + 1

            if memo[i][j] > max_length:
                max_length = memo[i][j]
                end_index = i   # track end position in s1
        else:
            memo[i][j] = 0

        return memo[i][j]

    # Fill memo table
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            solve(i, j)

    common_substring = s1[end_index - max_length:end_index]

    return common_substring, max_length


if __name__ == "__main__":
    file_path = r"C:\Users\narni\OneDrive\Desktop\sem-4 projects\daa\lcs_50_100.xlsx"

    # Read Excel safely
    df = pd.read_excel(file_path, header=None)
    df = df.dropna()

    strings = df.iloc[:, 0].astype(str).tolist()

    if len(strings) < 2:
        raise ValueError("Excel must contain exactly two strings (two rows).")

    s1 = strings[0]
    s2 = strings[1]

    print("\n===== INPUT STRINGS =====")
    print("String 1:", s1)
    print("Length 1:", len(s1))
    print("String 2:", s2)
    print("Length 2:", len(s2))

    start = time.perf_counter()
    substring, length = lcs_substring_memo(s1, s2)
    end = time.perf_counter()

    print("\n===== RESULT =====")
    print("Longest Common Substring:", substring)
    print("Length of Longest Common Substring:", length)
    print("Time taken:", end - start, "seconds")
```
